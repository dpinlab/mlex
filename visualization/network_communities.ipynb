# Network Community Detection:

## Testing Girvan-Newman Algorithm with Networkx:

```python
G = nx.read_gml('I-d_results_data-2.0/new-data_all_5_2_2021/network_graph_phi.graphml')
nx.draw(G)
```

```python
nx.edge_betweenness_centrality(G=G)
```

```python
NUM_ITER = 3
for i in range(NUM_ITER):
    edge_betweenness = nx.edge_betweenness_centrality(G=G).items()
    edge_to_delete = sorted(edge_betweenness,key=lambda pair: -pair[1])[0][0]
    G.remove_edge(*edge_to_delete)
    nx.draw(G=G,node_color='blue')
    plt.title('Step %s\nEdge %s Deleted'%(i,edge_to_delete), fontsize=20)
    plt.show()
```

## CDLIB API Girvan-Newman Algorithm:

### 2019 I-d Network:

```python
import cdlib.algorithms
from cdlib.viz import plot_network_clusters
G_network_2019 = nx.read_gml('I-d_results_data-2.0/new-data_high_5_2_2019/network_graph_phi.graphml')
partition_2019 = cdlib.algorithms.girvan_newman(g_original=G_network_2019, level=1)
plot_network_clusters(graph=G_network_2019, partition=partition_2019)
```

### 2020 I-d Network:

```python
G_network_2020 = nx.read_gml('I-d_results_data-2.0/new-data_high_5_2_2020/network_graph_phi.graphml')
partition_2020 = cdlib.algorithms.girvan_newman(g_original=G_network_2020, level=1)
plot_network_clusters(graph=G_network_2020, partition=partition_2020)
```
