```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from create_rr_phi_network import DataPreparer
```

```python
df = DataPreparer.read_data(r"/data/pcpe")
df2 = DataPreparer.read_final_typology_label(r"/data/pcpe")

df['typo'] = df[['I-d', 'I-e', 'IV-n']].agg(''.join, axis=1).map({
    '000': '',
    '100': 'I-d',
    '010': 'I-e',
    '001': 'IV-n',
    '101': 'I-d_IV-n'
})

df2.columns = ['CONTA_TITULAR', 'I-d_FINAL', 'I-e_FINAL', 'IV-n_FINAL']

df3 = pd.merge(df, df2, on="CONTA_TITULAR", how='outer')
```

```python
df3['any_typo'] = df3[['I-d', 'I-e', 'IV-n']].any(axis=1).astype(int).astype(str)
df3['any_typo_FINAL'] = df3[['I-d_FINAL', 'I-e_FINAL', 'IV-n_FINAL']].any(axis=1).astype(int).astype(str)
df3.replace(";", "-", regex=True, inplace=True)
df3['transactions'] = 1
```

```python
df3.to_csv("/data/pcpe/william_pcpe.csv", sep=";", index=False)
```
