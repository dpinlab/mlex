# LaTeX Export Example for PCPEDataExploratoryAnalysis

This notebook demonstrates how to use the `export_all_analyses_latex` method to generate LaTeX tables for all analyses.


```python
%load_ext autoreload
%autoreload 2
```

```python
import sys
from pathlib import Path

# Add the project root to Python path for Jupyter notebooks
project_root = Path.cwd().parent
sys.path.append(str(project_root))
```

```python
# Import required libraries
import pandas as pd
import numpy as np
from mlex import PCPEDataExploratoryAnalysis, DataReader, get_pcpe_dtype_dict, pcpe_preprocessing_read_func
import os

# Set up display options for better output
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
data_path = r'/data/pcpe/pcpe_04.csv'
target_column = 'I-d'

reader = DataReader(data_path, target_columns=[target_column], dtype_dict=get_pcpe_dtype_dict(), preprocessing_func=pcpe_preprocessing_read_func)
df = reader.read_df()
print(f"Sample dataset created with {len(df)} transactions")
print(f"Columns: {list(df.columns)}")
df.head()
```

```python
# Initialize the PCPEDataExploratoryAnalysis class
analyzer = PCPEDataExploratoryAnalysis(df)
```

```python
# Export all analyses to LaTeX tables
# This will create individual .tex files for each analysis

latex_tables = analyzer.export_all_analyses_latex(
    output_dir="latex_tables",           # Directory to save .tex files
    top_n=10,                           # Number of top items in ranking tables
    typology_columns=['I-d']  # Specific typologies to analyze
)

print(f"\nGenerated {len(latex_tables)} LaTeX tables:")
for name in latex_tables.keys():
    print(f"  - {name}.tex")
```

```python
# Display the content of one of the generated LaTeX tables
print("Example LaTeX table content (Bank Analysis):")
print("=" * 60)
print(latex_tables['bank_analysis'])
```

```python
# Check the generated files
if os.path.exists("latex_tables"):
    files = os.listdir("latex_tables")
    print(f"Generated files in 'latex_tables' directory:")
    for file in sorted(files):
        if file.endswith('.tex'):
            file_path = os.path.join("latex_tables", file)
            file_size = os.path.getsize(file_path)
            print(f"  - {file} ({file_size} bytes)")
else:
    print("No latex_tables directory found")
```

```python
# Generate a complete LaTeX document with all analyses
latex_document = analyzer.generate_latex_document(
    output_file="complete_analysis_report.tex",
    title="Financial Transaction Analysis Report",
    author="Data Analysis Team",
    top_n=10,
    typology_columns=['I-d', 'Risk_Level', 'Suspicious_Flag']
)

print(f"Complete LaTeX document generated: {latex_document}")

# Check if the document was created
if os.path.exists("complete_analysis_report.tex"):
    file_size = os.path.getsize("complete_analysis_report.tex")
    print(f"Document size: {file_size} bytes")
    
    # Show first few lines of the document
    with open("complete_analysis_report.tex", 'r', encoding='utf-8') as f:
        lines = f.readlines()[:20]  # First 20 lines
        print("\nFirst 20 lines of the LaTeX document:")
        print("-" * 50)
        for i, line in enumerate(lines, 1):
            print(f"{i:2d}: {line.rstrip()}")
```

```python
# Example: Export specific analyses with custom settings
print("Exporting specific analyses with custom settings:")
print("=" * 60)

# Export bank analysis with custom caption and top 15 banks
bank_latex = analyzer.export_bank_analysis_latex(
    top_n=15,
    caption="Top 15 Banks by Transaction Volume and Distribution",
    label="tab:top_banks_detailed"
)
print("Bank Analysis LaTeX:")
print(bank_latex)
print("\n" + "="*60 + "\n")

# Export typology analysis with custom settings
typology_latex = analyzer.export_typology_analysis_latex(
    typology_column='Risk_Level',
    typology_name='Risk Assessment',
    caption="Risk Level Distribution Analysis",
    label="tab:risk_analysis"
)
print("Risk Level Typology Analysis LaTeX:")
print(typology_latex)
```

```python
# Save individual LaTeX tables to files with custom names
custom_tables = {
    'bank_analysis_custom.tex': analyzer.export_bank_analysis_latex(top_n=20),
    'risk_analysis_custom.tex': analyzer.export_typology_analysis_latex('Risk_Level'),
    'missing_data_custom.tex': analyzer.export_missing_data_analysis_latex()
}

# Create custom directory
custom_dir = "custom_latex_tables"
os.makedirs(custom_dir, exist_ok=True)

# Save custom tables
for filename, latex_content in custom_tables.items():
    filepath = os.path.join(custom_dir, filename)
    with open(filepath, 'w', encoding='utf-8') as f:
        f.write(latex_content)
    print(f"Saved: {filepath}")

print(f"\nCustom LaTeX tables saved in '{custom_dir}' directory")
```

## Summary

This example demonstrates:

1. **Basic Usage**: Using `export_all_analyses_latex()` to generate all analysis tables
2. **Custom Settings**: Specifying output directory, top N items, and typology columns
3. **Individual Exports**: Exporting specific analyses with custom captions and labels
4. **Complete Document**: Generating a full LaTeX document with all analyses
5. **File Management**: Organizing and saving LaTeX files in custom directories

The generated LaTeX files can be:
- Included in your LaTeX documents using `\input{filename.tex}`
- Compiled directly with `pdflatex`
- Used in academic papers, reports, or presentations
